```python
#lab 9 196 task 1 
import cv2  
import matplotlib.pyplot as plt  
import numpy as np  

# Load the uploaded image
image = cv2.imread('img.1.196.jpg')  # Use the uploaded image

if image is not None:
    # Convert the image from BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

    # Display the original image 
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
    plt.imshow(image_rgb)  
    plt.title('Original Image')  
    plt.axis('off')  # Hide axis  

    # Increase brightness
    brightness_increase = 50  # You can change this value
    bright_image = cv2.convertScaleAbs(image, alpha=1, beta=brightness_increase)

    # Convert the bright image to RGB for display
    bright_image_rgb = cv2.cvtColor(bright_image, cv2.COLOR_BGR2RGB)

    # Display the augmented image
    plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
    plt.imshow(bright_image_rgb)  
    plt.title('Brightened Image')  
    plt.axis('off')  
    plt.show()  
else:
    print("Error: Image not found. Please check the path.")
```

```python
#lab 9 196 task 2
import cv2  
import matplotlib.pyplot as plt  
import numpy as np  

# Load the uploaded image
image = cv2.imread('img.1.196.jpg')  # Use the uploaded image

if image is not None:
    # Convert the image from BGR to RGB for display purposes
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)  

    # Apply Sobel edge detection
    sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)  # Sobel in X direction
    sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5)  # Sobel in Y direction
    sobel_edge = cv2.magnitude(sobel_x, sobel_y)  # Combine both directions

    # Display the original grayscale image and the Sobel edge-detected image
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
    plt.imshow(gray_image, cmap='gray')  
    plt.title('Grayscale Image')  
    plt.axis('off')  

    plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
    plt.imshow(sobel_edge, cmap='gray')  
    plt.title('Sobel Edge Detection')  
    plt.axis('off')  

    plt.show()  
else:
    print("Error: Image not found. Please check the path.")
```
